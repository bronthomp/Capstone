```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
mortality = pd.read_csv(r'C:\Users\brthompson\bt_school\data\interim\firstmodel_mortality.csv')
readmit = pd.read_csv(r'C:\Users\brthompson\bt_school\data\interim\firstmodel_readmit.csv')
```

```python
##First employ chi-square feature selection
#start with readmission classification model
readmit = readmit.drop(['HOSPITAL_EXPIRE_FLAG'], 1)
readmit.shape
```

```python
array = readmit.values
Xad = array[:, 1:]
Yad = array[:,0]
```

```python
tester = SelectKBest(score_func=chi2, k=4)
fit = tester.fit(Xad, Yad)
```

```python
np.set_printoptions(precision=3)
print(fit.scores_)
```

```python
#In order to interpret scores, append them to a dataframe with the column ames
readmit_chisq= pd.DataFrame()
readmit_chisq['scores'] = fit.scores_
readmit_chisq['names'] = readmit.columns[1:]
```

```python
readmit_chisq.sort_values(by = 'scores', ascending = False)
```

```python
readmit_chisq['scores'].mean()
```

```python
readmit_chisq['scores'].median()
```

```python
keep_readmit_chisq = readmit_chisq[readmit_chisq['scores']>11]
```

```python
keep_readmit_chisq.shape
```

```python
#Repeat for mortality outcome 
mort = mortality.drop(['readmit'], 1)
mort.shape
```

```python
array1 = mort.values
Xm = array1[:, 1:]
Ym = array1[:,0]
```

```python
tester = SelectKBest(score_func=chi2, k=4)
fit = tester.fit(Xm, Ym)
```

```python
np.set_printoptions(precision=3)
print(fit.scores_)
```

```python
mort_chisq= pd.DataFrame()
mort_chisq['scores'] = fit.scores_
mort_chisq['names'] = mort.columns[1:]
```

```python
mort_chisq.sort_values(by = 'scores', ascending=False)
```

```python
mort_chisq['scores'].mean()
```

```python
mort_chisq['scores'].median()
```

```python
#Keep those that are over the median score (round to 3)
keep_mort_chisq = mort_chisq[mort_chisq['scores']>3]
keep_mort_chisq.shape
```

```python
##Next use Recursive Feature Elimination for each outcome of interest
model = LogisticRegression()
rfe = RFE(model, 100)
fit = rfe.fit(Xad, Yad)
print("Num Features: %s" % (fit.n_features_))
print("Selected Features: %s" % (fit.support_))
print("Feature Ranking: %s" % (fit.ranking_))
```

```python
len(fit.support_)
```

```python
len(readmit.columns)
```

```python
readmit_rfe = pd.DataFrame()
readmit_rfe['include'] = fit.support_
readmit_rfe['names'] = readmit.columns[1:]
readmit_rfe
```

```python
keep_readmit_rfe = readmit_rfe[readmit_rfe['include']==True]
keep_readmit_rfe['names']
```

```python
##Repeat for mortality outcome
model = LogisticRegression()
rfe = RFE(model, 100)
fit = rfe.fit(Xm, Ym)
print("Num Features: %s" % (fit.n_features_))
print("Selected Features: %s" % (fit.support_))
print("Feature Ranking: %s" % (fit.ranking_))
```

```python
len(fit.support_)
```

```python
len(mort.columns)
```

```python
mort_rfe = pd.DataFrame()
mort_rfe['include'] = fit.support_
mort_rfe['names'] = mort.columns[1:]
mort_rfe
```

```python
keep_mort_rfe = mort_rfe[mort_rfe['include']==True]
keep_mort_rfe['names']
```

```python
#Choose features that are in top 100 of either one
readmit_selected = (list(keep_readmit_chisq['names']) + list(keep_readmit_rfe['names']))
```

```python
readmit_selected.append('readmit')
```

```python
mort_selected = (list(keep_mort_chisq['names']) + list(keep_mort_chisq['names']))
```

```python
mort_selected.append('HOSPITAL_EXPIRE_FLAG')
```

```python
readmit_model1 = readmit[readmit_selected]
```

```python
mort_model1 = mort[mort_selected]
```

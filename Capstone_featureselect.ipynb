```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r'C:\Users\brthompson\bt_school\data\interim\firstmodel.csv')
```

```python
##First employ chi-square feature selection
#start with readmission classification model
readmit = df.drop(['HOSPITAL_EXPIRE_FLAG'], 1)
readmit.dtypes
```

```python
array = readmit.values
Xad = array[:, 1:]
Yad = array[:,0]
```

```python
tester = SelectKBest(score_func=chi2, k=4)
fit = tester.fit(Xad, Yad)
```

```python
np.set_printoptions(precision=3)
print(fit.scores_)
```

```python
#In order to interpret scores, append them to a dataframe with the column ames
temp= pd.DataFrame()
temp['scores'] = fit.scores_
temp['names'] = readmit.columns[1:]
```

```python
temp.sort_values(by = 'scores', ascending = False)
```

```python
#Repeat for mortality outcome 
mort = df.drop(['readmit'], 1)
mort.dtypes
```

```python
array1 = mort.values
Xm = array1[:, 1:]
Ym = array1[:,0]
```

```python
tester = SelectKBest(score_func=chi2, k=4)
fit = tester.fit(Xm, Ym)
```

```python
np.set_printoptions(precision=3)
print(fit.scores_)
```

```python
temp1= pd.DataFrame()
temp1['scores'] = fit.scores_
temp1['names'] = mort.columns[1:]
```

```python
temp1.sort_values(by = 'scores', ascending=False)
#Should actually exclude discharge location from this one since it is an indicator of death which is the outcome we are trying to predict
```

```python
##Next use Recursive Feature Elimination for each outcome of interest
model = LogisticRegression()
rfe = RFE(model, 30)
fit = rfe.fit(Xad, Yad)
print("Num Features: %s" % (fit.n_features_))
print("Selected Features: %s" % (fit.support_))
print("Feature Ranking: %s" % (fit.ranking_))
```

```python
fit.support_[1]
```

```python
readmit_rfe = pd.DataFrame()
readmit_rfe['include'] = fit.support_
readmit_rfe['names'] = readmit.columns[1:]
readmit_rfe
```

```python
keep_readmit = readmit_rfe[readmit_rfe['include']==True]
keep_readmit['names']
```

```python
##Repeat for mortality outcome
model = LogisticRegression()
rfe = RFE(model, 30)
fit = rfe.fit(Xm, Ym)
print("Num Features: %s" % (fit.n_features_))
print("Selected Features: %s" % (fit.support_))
print("Feature Ranking: %s" % (fit.ranking_))
```
